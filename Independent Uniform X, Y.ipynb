```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
n=0
for i in range(1000):
    X=np.random.uniform(0,1)
    Y=np.random.uniform(0,1)
    if( X**2 + Y**3 > 0.75 ):
        n+=1
print("probability =",(n/1000))
```

```python
n=0
for i in range(1000):
    X=np.random.uniform(0,1)
    Y=np.random.uniform(0,1)
    if( X + Y > 5 * X  * (Y**(0.5))):
        n+=1
print("probability =",(n/1000))
```

```python
def f(x):
    return (math.exp(x**2/2))
a=-1
b=1
sum=0
x=np.linspace(0,1000,1000)
y=[]
for i in range (1000):
    Xn=np.random.uniform(-1,1)
    sum+=f(Xn)
    y.append((b-a)*sum/(i+1))
plt.plot(x,y)
plt.ylim()
plt.ylim(2,3)
plt.title("Monte Carlo")
plt.xlabel("num of samples")
plt.ylabel("value of integral")
plt.show()
I=(b-a)*sum/1000
print("I =",I)
```
