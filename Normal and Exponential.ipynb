```python
import numpy as np
import matplotlib.pyplot as plt

def exp_inverse(Fx):
    x=-np.log(1-Fx)
    return x

new_exp=[]

for i in range (100):
    Fx=np.random.uniform(0,1)
    x_exp=exp_inverse(Fx)
    new_exp.append(x_exp)
    
plt.hist(new_exp,bins=10)
plt.title("exponential distribution")
plt.show()
```

```python
def norm_inverse(Fx):
    from scipy.stats import norm
    x=norm.ppf(Fx)
    return x

new_norm=[]

for i in range (100):
    Fx=np.random.uniform(0,1)
    x_norm=norm_inverse(Fx)
    new_norm.append(x_norm)
    
plt.figure()
plt.hist(new_norm,bins=10)
plt.title("normal distribution")
plt.show()
```
