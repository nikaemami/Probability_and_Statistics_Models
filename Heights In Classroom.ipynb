**Mathematical View:**

```python
import math
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(2,61,1)
probability = []

i = 1
while (i < 60):
    probability.append(1-(math.factorial(320)/((math.factorial(320-(i+1)))*(320**(i+1)))))
    i=i+1
    
y=np.array(probability)

plt.plot(x,y)
plt.title('height formula')
plt.xlabel('num of people in the class')
plt.ylabel('probability')
plt.show()

for j in range (58):
    if probability[j+1]>=0.5 and probability[j]<0.5:
        print('min number of people: ',j+3)
```

**Programming View:**

```python
import numpy as np
import matplotlib.pyplot as plt
import random

x=np.arange(2,61,1)
probability=[]

for i in range(59):
    favorable=0
    for j in range(99):
        classes=[]
        for k in range (i+2):
            classes.append(random.randint(1,320))
        classes.sort()
        for k in range (i+1):
            if(classes[k]==classes[k+1]):
                favorable+=1
                break       
    probability.append(favorable/100)
    
y=np.array(probability)

plt.plot(x,y)
plt.title('height practical')
plt.xlabel("num of people in the class")
plt.ylabel("probability by favorable classes")
plt.show()
```
