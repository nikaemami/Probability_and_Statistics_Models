```python
import numpy as np
import matplotlib.pyplot as plt
import math
n=1
ep=0
sum_ep=0
epsilon=[]
while (n<=500):
    sum_ep=0
    r=1
    while(r<=n):
        x=float(n/r)
        y=math.ceil(x)
        ep=y-x
        sum_ep+=ep
        r+=1
    if(n==1):
        epsilon.append(0)
    else:
        epsilon.append(float(sum_ep/(n-1)))
    n+=1
x=np.linspace(1,500,500)
plt.plot(x,epsilon)
plt.title("expected value")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
epsilon1=[]
epsilon2=[]
n=2
while (n<=500):
    r1=np.random.randint(1,n)
    x=float(n/r1)
    y=math.ceil(x)
    ep1=y-x
    epsilon1.append(ep1)
    
    r2=np.random.randint(1,n+1)
    x2=float((n+1)/r2)
    y2=math.ceil(x2)
    ep2=y2-x2
    epsilon2.append(ep2)
    
    n+=1

delta=[]
for i in range (499):
    delta.append((epsilon2[i]-0.5772)/(epsilon1[i]-0.5772))

x=np.linspace(2,500,499)
plt.plot(x,delta)
plt.title('value of delta')
plt.show()
```
